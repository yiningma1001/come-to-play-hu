```python
from mpl_toolkits.basemap import Basemap
import pandas as pd
import csv
import xlrd
import matplotlib.pyplot as plot
from pandas import Series
from matplotlib.patches import Polygon

##州统计献金
def state(data):
    state_vs = data.groupby(['committee_name','contributor_state'])['contributor_aggregate_ytd'].sum().unstack(level = 0)
    state_vs.fillna(0, inplace=True)
    plot.figure(figsize=(23,9))
    ax = plot.subplot(1,1,1)
    state_vs.plot(kind = 'bar',ax = ax)
    state_vs_rate = state_vs.div(state_vs.sum(axis=1), axis=0)
    state_vs_rate.drop(labels=['AA','AE'],inplace = True)
    print(state_vs_rate)
    biden = state_vs_rate['BIDEN FOR PRESIDENT']
    abbr = {'Commonwealth of Kentucky':'KY','Commonwealth of Massachusetts':'MA','Commonwealth of Pennsylvania':'PA',
             'State of Rhode Island and Providence Plantations':'RI'}
    from matplotlib.colors import rgb2hex

    plot.figure(figsize=(12,9))
    # lat_1 or lat_0 and lon_0
    m = Basemap(llcrnrlon=-130, llcrnrlat=20, urcrnrlon=-65, urcrnrlat=49,
                projection = 'lcc',
                lat_1 =30,
                lon_0 = -100
                )

    m.drawcoastlines(linewidth=1.5)
    m.drawcountries(linewidth = 1.5)
# m.drawstates()

# 地图读取了美国地图形状，m中就有各州的形状，数据
    m.readshapefile('/Users/mayining/Downloads/gadm36_USA_shp/gadm36_USA_1',name = 'states')


    colors = []
    states = []

    cmap = plot.cm.Reds
    cmap1=plot.cm.Blues

    for shapeinfo in m.states_info:
        a = shapeinfo['VARNAME_1']
#     州的缩写
        s = a.split('|')[0]
    
#     可以代表颜色
        try:
            rate = biden[s]
            if(rate>0.5):
                colors.append(cmap(rate))
            else:
                colors.append(cmap1(1-rate))
            states.append(s)
            
        except:
            colors.append(cmap(biden[abbr[s]]))
            if(rate>0.5):
                colors.append(cmap(biden[abbr[s]]))
            else:
                colors.append(cmap1(biden[abbr[s]]))
            states.append(s)
    
# 州填充颜色
# seg州中的一部分区域，多边形
    ax = plot.gca()
    for n,seg in enumerate(m.states):
        c = rgb2hex(colors[n])
        poly = Polygon(seg,color = c )
        ax.add_patch(poly)
    
    plot.show()




filename='/Users/mayining/PycharmProjects/pythonProject3/total_2020.csv'
data=pd.read_csv(filename)
state(data)
```
